```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load model and tokenizer
model_name = "gpt2"  # you can also try 'distilgpt2' for a smaller version
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
model.eval()

def generate_poem(topic, max_length=100):
    prompt = f"Write a short poem about {topic}:\n"
    inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model.generate(
            inputs,
            max_length=max_length,
            num_return_sequences=1,
            do_sample=True,
            temperature=0.8,
            top_k=50,
            top_p=0.95,
            eos_token_id=tokenizer.eos_token_id
        )    

    generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
    poem = generated[len(prompt):].strip()
    return poem
```

```python
while True:
    user_input = input("Enter a topic for the poem (or 'quit' to exit): ")
    if user_input.lower() == 'quit':
        print("Goodbye! ✨")
        break
    poem = generate_poem(user_input)
    print("\nHere's your poem:\n")
    print(poem)
    print("\n" + "-"*50 + "\n")
```
